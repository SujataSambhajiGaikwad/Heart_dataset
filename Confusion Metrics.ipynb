## Balancing data

```python
from imblearn.over_sampling import SMOTE
s1=SMOTE()
x_data1,y_data1=s1.fit_resample(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_data1,y_data1,test_size=0.2,random_state=11)
```

## cross - validation technique

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=5,random_state=11,shuffle=True)
kf.get_n_splits(x)
print(kf)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.linear_model import LogisticRegression 
from collections import Counter
l1=LogisticRegression()
```

```python
for train_data,test_data in kf.split(x):
    x_train,x_test=x_data1[train_data],x_data1[test_data]
    y_train,y_test=y_data1[train_data],y_data1[test_data]
    scores=cross_val_score(l1,x_train,y_train,cv=kf)
    y_pred=cross_val_predict(l1,x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print(confusion_matrix(y_pred,y_test))
```
